# Continuous Experiments at Uber
#### Jeremy Gu - Senior Data Scientist @ Uber

- Methodology divided into two parts:
    - Randomized Experiments
        - Classic Experiments (Non-recurring)
            - Univariate Tests (A/B Tests, Adaptive A/B)
            - Multivariate Tests (Factorial Designs)
        - Continuous Experiments (Recurring)
            - Heuristic Techniques
            - Model Based Techniques
    - Observational Studies
        - Difference in Difference
        - Synthetic Control
        - Multivariate Matching
   
   
- Business Use Cases at Uber:
    - Product Feature - Evaluating how feature will impact experience
    - Marketing Message - What kind of message resonates with drivers and riders
    - Incentive Spend 
    - Monitoring: Is driver following our route? Which way is safer? (To check this, we can't randomize). So here have to use observational study.
- Continous experiments allow 'real time learning'


- "One Arm Bandit" slot machine use cases in Uber:
    - Content Optimization
    - Hyper parameter Tuning
    - Spend Optimization
    
    
- Uber Eats Ranking Algorithm:
    - BUSINESS PROBLEM: Rank the restaurants to drive more revenue, higher conversion, and higher retention rates.
    - Modelling Components:
        - Linear Programming; An optimization problem with multiple objectives
        - Machine Learning: Prediction of Business Metrics
    - Statistical Methdods: Continuous Experiments
    



##### Mandie Liu:

- Staged-Rollout
- Any code change to the app: Front End or Back End (How do any of these changes make an impact?)
- When do you want to roll out the feature? You want the feature to benefit many people, but what if people don't like it? To solve it:
    - Sequential Testing: Conduct experiment between treatment and control group
        - Power Based and Time Series Models
    - Go from 1%, 5%, 20%, 50% to 100% in treatment sample. Go to 100% treatment sample, and control sample = 0%. 
    - If there is degredation, roll back the feature!
    - If no degredation: Can be case that the feature may not be liked, but we don't have enough data to figure that out.

# Machine Learning on the Blockchain

##### Kyle Polich - Data Skeptic Host
@dataskeptic
dataskeptic.com

- Parallelization has been a big help to ML.
- Algorithms aren't always the key, data are. 
- Data often has non-linear relationships, so linear models don't often cut it.
- Another problem is: data is often high dimensional.

- How do we find areas that are hard, and make them easy?
    - Data Parallelism: Hadoop, Spark etc. 
    - Spark Streaming, Serverless Architecture
    
    
- Where will ML go?
    - Mobiles: Get TensorFlow to run on mobiles.
    - BlockChain: You can reverse engineer models by finding them from the Cloud
    - Can we protect our models someway? "Data is the new oil."



- Spam Filtering
    - Gmail saw trends in spam data.
    - How do people play games? => Big video game companies will have access to this data, smaller game companies might not.
    
    
    
- Opting in/Out: "I don't want my data in your model."
    - I want to be in, but I want to be anonymous.
    - I want to give you my data, but I want something in return (tips/advice/forecasts)
    
    
- Anonymity:
    - "Tell me how many of these activities have you done?"
    - Sensitive questions: will people be less likely to respond to these?
    - Just because you are anonymous doesn't mean you don't stand out.
    
    
- Homomorphic Encryption
- Benefits of Blockchain:
    - Decentralized, Immutability, High availability, Highly secure, Shared Ledger, Transparency & Trust
    
    
- Google Research's Federated Learning
- Algorithmia
- Singularity


- "I want them to have my data, but I don't want them to do other things with it." 
    - What if they only get the bits that help improve their model to give me recommendations, and don't take other data?
    - "Trust, but verify."
    - Data Provenance via blockchain: Where did data come from? Who's seen it? Who's using it?

# Data Science in Risk, Economic, and Financial Asset Price Forecast

##### Dr. Yi (Clara) Liu â€“ Manager, Data & Analytics from KPMG

- Most banks are required to make forecasts for how likely customer will default on their loan, and how much of it can be recovered if the customer does default.

- Classification problem: Either loan is defaulted or not. 
- In most cases, not a linear relationship between explanatory and explained variables. 


- Random forest, Gradient Boosting, SVMs helpful for non-linear models:
    - Difficult to understand and interpret, especially with impact of each feature.
    
    
- Partial Dependence Plot:
    - Relationship between feature and target after controlling for the effects of other variables.
    
    
- Difficult to quantify marginal effect of each feature in Logistic Regression
        
- LSTM models are a challenge because of the nature of the loss function.
    - Model Tuning:
        - Track loss function closely to avoid over-fitting
        - Early Stopping: Stop training as soon as validation error starts to increase.
